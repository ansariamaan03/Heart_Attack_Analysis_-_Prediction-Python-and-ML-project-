#Introduction -
A heart attack, also called a myocardial infarction, happens when a part of the heart muscle doesn't get enough blood.
The more time that passes without treatment to restore blood flow, the greater the damage to the heart muscle.
Coronary artery disease (CAD) is the main cause of heart attack.

# Outlier Detection

```python
numeric_list = ["age", "trtbps","chol","thalachh","oldpeak"]
df_numeric = df.loc[:, numeric_list]
df_numeric.head()
```

```python
for i in numeric_list:

    # IQR
    Q1 = np.percentile(df.loc[:, i],25)
    Q3 = np.percentile(df.loc[:, i],75)

    IQR = Q3 - Q1

    print("Old shape: ", df.loc[:, i].shape)

    # upper bound
    upper = np.where(df.loc[:, i] >= (Q3 +2.5*IQR))

    # lower bound
    lower = np.where(df.loc[:, i] <= (Q1 - 2.5*IQR))

    print("{} -- {}".format(upper, lower))

    try:
        df.drop(upper[0], inplace = True)
    except: print("KeyError: {} not found in axis".format(upper[0]))

    try:
        df.drop(lower[0], inplace = True)
    except:  print("KeyError: {} not found in axis".format(lower[0]))

    print("New shape: ", df.shape)
```

#Modelling

```python
df1 = df.copy()
```

```python
df1 = pd.get_dummies(df1, columns = categorical_list[:-1], drop_first = True)
df1.head()
```

```python
x = df1.drop(["output"], axis = 1)
y = df1[["output"]]
```

```python
x[numeric_list[:-1]] = scaler.fit_transform(x[numeric_list[:-1]])
x.head()
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.1, random_state = 3)
print("X_train: {}".format(xtrain.shape))
print("X_test: {}".format(xtest.shape))
print("y_train: {}".format(ytrain.shape))
print("y_test: {}".format(ytest.shape))
```

Logistic Regression - Logistic regression is a supervised machine learning algorithm used for classification tasks where the goal is to predict the probability that an instance belongs to a given class or not. Logistic regression is a statistical algorithm which analyze the relationship between two data factors.

```python
logreg = LogisticRegression()
logreg
```

```python
logreg.fit(xtrain, ytrain)
```

```python
y_pred_prob = logreg.predict_proba(xtest)
y_pred_prob
```

```python
y_pred = np.argmax(y_pred_prob, axis = 1)
y_pred
```

```python
print("Test accuracy: {}".format(accuracy_score(y_pred, ytest)))
```

```python
# ROC Curve
fpr, tpr, thresholds = roc_curve(ytest, y_pred_prob[:,1])
```

```python
plt.plot([0,1],[0,1],"k--")
plt.plot(fpr, tpr, label = "Logistic Regression")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Logistic Regression ROC Curve")
plt.show()
```

```python
lr = LogisticRegression()
lr
```

```python
penalty = ["l1", "l2"]
parameters = {"penalty":penalty}
```

```python
lr_searcher = GridSearchCV(lr, parameters)
```

```python
lr_searcher.fit(xtrain, ytrain)
```

```python
print("Best parameters: ",lr_searcher.best_params_)
```

```python
y_pred = lr_searcher.predict(xtest)
```

```python
print("Test accuracy: {}".format(accuracy_score(y_pred, ytest)))
```

From these results, you can draw conclusions about the effectiveness of the logistic regression model in predicting heart disease. If not, further tuning or the use of more advanced models may be necessary to improve performance.


# Decision Tree


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
x = df.drop('output', axis=1)
y = df['output']
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Initialize the Decision Tree model
model = DecisionTreeClassifier(random_state=42)
```

```python
# Train the model
model.fit(xtrain, ytrain)
```

```python
# Make predictions
y_pred = model.predict(xtest)
```

```python
# Evaluate the model
accuracy = accuracy_score(ytest, y_pred)
conf_matrix = confusion_matrix(ytest, y_pred)
class_report = classification_report(ytest, y_pred)
```

```python
# Display the results
print(f"Accuracy: {accuracy * 100:.2f}%")
print("Confusion Matrix:")
print(conf_matrix)
print("Classification Report:")
print(class_report)
```

```python
# Plot the decision tree
plt.figure(figsize=(20,10))
tree.plot_tree(model, feature_names=x.columns, class_names=['No Heart Disease', 'Heart Disease'], filled=True, rounded=True)
plt.show()
```

# Conclusion of Decision Tree:
The Decision Tree model provides a good balance between accuracy and interpretability. With an accuracy of around 82%, it shows reliable performance in predicting heart disease.
The low number of false negatives and false positives indicates that the model is reasonably precise and sensitive.
The classification report confirms that the model performs well for both classes (presence and absence of heart disease), with high precision and recall.
The decision tree plot allows for easy interpretation of the model, showing how different features contribute to the prediction.

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

x = df.drop('output', axis=1)
y = df['output']

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Initialize the Random Forest model
model = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
# Train the model
model.fit(xtrain, ytrain)
```

```python
# Make predictions
y_pred = model.predict(xtest)
```

```python
# Evaluate the model
accuracy = accuracy_score(ytest, y_pred)
conf_matrix = confusion_matrix(ytest, y_pred)
class_report = classification_report(ytest, y_pred)
```

```python
# Display the results
print(f"Accuracy: {accuracy * 100:.2f}%")
print("Confusion Matrix:")
print(conf_matrix)
print("Classification Report:")
print(class_report)
```

#Conclusion:
The Random Forest classifier shows strong potential for heart attack prediction with an overall accuracy of 85%. It is particularly effective in correctly identifying heart attack cases, with a high recall rate of 0.92, indicating that it captures the majority of true heart attack cases. However, the presence of false negatives highlights the need for further model refinement to reduce missed diagnoses.

In a clinical context, the model could be a valuable decision-support tool, aiding healthcare professionals in identifying at-risk patients for further diagnostic evaluation and intervention. Further improvements could involve tuning the model parameters, incorporating more features, or using additional data to enhance predictive performance, particularly in minimizing false negatives.

Overall, while the model is a promising tool for heart attack prediction, ongoing validation and refinement are essential to ensure it meets the high standards required for clinicalapplication.
